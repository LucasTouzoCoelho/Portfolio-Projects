## 3rd Step - Exploratory Analysis

#### 1) Finished the data cleaning, let's se a big picture of our features:

```python
df.info()
```

Nice! Now, we can start our exploratory analysis! Let's begin seeing the data distribution

```python
df.head()
```

So, we're gonna start creating a new column to put the age column in age groups. 

```python
faixas =['1-10','11-20','21-30','31-40','41-50','51-60','+60']
```

```python
df["paciente_faixa_etaria"] = pd.cut(x=df['paciente_idade'], bins=[0,10,20,30,40,50,60,150], labels=faixas)
```

```python
df.head()
```

So, let's check the vaccination distribution per age group:

#### 2) What is the data distribution if we separate by age groups?

```python
sns.displot(data=df['paciente_faixa_etaria'], x=df['paciente_faixa_etaria']).set(title="Age Groups")
plt.show()
```

Seeing the visualization above, we realize that the most of people that were vaccinated against COVID-19 were in the +60 age. In Brazil, the elder group has the priority to receive the vaccine, and mainly in the beggining of the vaccination, these risk groups had a great accession.
The second more vaccinated group was the 31-40 age, and here we have most of the health professionals being vaccinated. They also are in the risk group.

Now we're gonna see if the race distribution in our sample:

#### 2) And the races? Does it have some influence in the vaccination?

```python
paciente_racaCor_cont = df['paciente_racaCor_valor'].value_counts()
fig, ax = plt.subplots(figsize=(15, 5))
paciente_racaCor_cont.plot(kind='bar', color = ['seagreen', 'mediumaquamarine'])
plt.title('Race Distribution',fontsize=25)
plt.xticks(rotation=0)
plt.show()
```

Seeing the race distribution, we see a big difference between the access of the White Race when we compare with other ones. We need to investigate more about this question to identify the reason of this. We can also investigate the why a big part of the people didn't declare their races.

Now, let's see the top cities that received the vaccines amount:

#### 3) And what about the cities? Is there any city that is highlighting?

```python
top_municipios = df['paciente_endereco_nmMunicipio'].value_counts()[0:10]
fig, ax = plt.subplots(figsize=(15, 5))
plt.title('Cities Distribution',fontsize=25)
top_municipios.plot(kind='bar', color = ['red', 'blue'])
plt.xticks(rotation=45)
plt.show()
```

Now we can realize that there is a big disparity between the cities. Sao Paulo was the city that received the biggest part of the amount of the vaccines, and there are some hypotheses:

- It was a logistic strategy, seeing that Sao Paulo was the place where the vaccines was being produced.
- Sao Paulo is the most populous place in the Latin America, so the impact is too much better when you can achieve many people.
- Sao Paulo's Government was the first institution in Brazil to buy vaccines from China, and seeing the urgency of the situation, took the a big part of the amount bought.

We're gonna see if we've a different between the different genders (wich are on the dataset)

#### 4) And between men and women? Is there a difference?

```python
paciente_enumSexoBiologico = df['paciente_enumSexoBiologico'].value_counts()
fig, ax = plt.subplots(figsize=(12, 5))
plt.title('Gender Distribution',fontsize=25)
sexo_paciente = paciente_enumSexoBiologico.plot(kind='bar', color = ['pink', 'cornflowerblue'])
plt.xticks(rotation=0)
plt.show()
```

There's a small difference between Male and Female people vaccinated. Is a attention point, but in a first momment it can't provide great conclusions for us.

So, let's check the vaccination distribution during the whole pandemic:

#### 5) Now let's look at the vaccination over the time?

```python
df['vacina_dataAplicacao'].hist(bins=15, figsize=(15,10))
plt.title('Vaccination during the pandemic',fontsize=25)
plt.show()
```

Looking to the visualization, we can realize that only around ten months after the first vaccine be applicated, the vaccination got on the highest level until 

```python
df_MF = df.loc[(df['paciente_enumSexoBiologico'] == 'F') 
                         | (df['paciente_enumSexoBiologico'] == 'M') ]
```

```python
percent_com = pd.crosstab(df.paciente_faixa_etaria , df_MF.paciente_enumSexoBiologico , normalize = 'index') * 100
percent_com
```

#### 6) Correlating the Age Groups and the gender

```python
# Plot

# Prepare the figure from the data
fig = percent_com.plot(kind = 'bar', 
                       figsize = (16, 8), 
                       width = 0.5, 
                       edgecolor = 'g', 
                       color = ['pink','b'])

# Draw each group
for i in fig .patches:
    fig.text(i.get_x() + 0.00, 
             i.get_height() + 0.3, 
             str(round((i.get_height()), 2)), 
             fontsize = 15, 
             color = 'black',
             rotation = 0)

# Write the title and print
plt.title("Age Groups vs Gender", fontsize = 15)
plt.show()
```

Here, we can realize that according we move on by the age groups, the difference between men and women increases. It could generate some insights about the reason why this happens.

Let's take a new analysis, seeing how the database behaves when we compare according the groups priority:

```python
percent_groups = pd.crosstab(df.vacina_categoria_nome , df_MF.paciente_enumSexoBiologico , normalize = 'index') * 100
percent_groups = percent_groups.sort_values(by=['M'],ascending=True)
percent_groups
```

#### 7) We can do the same to other features, for example, the priority groups:

```python
# Plot

# Prepare the figure from the data
fig = percent_groups.plot(kind = 'barh', 
                       figsize = (16, 15), 
                       width = 0.5, 
                       edgecolor = 'g', 
                       color = ['pink','b'])


# Draw each group
for i in fig .patches:
    fig.text(i.get_x() + 0.00, 
             i.get_height() + 0.3, 
             str(round((i.get_height()), 2)), 
             fontsize = 15, 
             color = 'black',
             rotation = 0)

# Write the title and print
plt.title("Priority Groups vs Gender", fontsize = 15)
plt.show()
```

#### 8) And did everyone received all the vaccine applications?

```python
df['vacina_descricao_dose'].hist(bins=15, figsize=(15,10))
plt.title('Vaccination during the pandemic',fontsize=25)
plt.show()
```

```python
df['vacina_categoria_nome'].value_counts()
```

We can see that, apart to the "Faixa etária", the group "Comorbidades" had the most important percent of applications. So let's analyse this especific group to se the vaccine distribution among the applications:

#### 9) In the main priority group, does everyone received all the vaccine applications?

```python
comorbidity = df[df['vacina_categoria_nome']=='Comorbidades']
comorbidity
```

```python
percent_app = pd.crosstab(comorbidity.vacina_categoria_nome , df.vacina_descricao_dose , normalize = 'index') * 100
percent_app = percent_app.sort_values(by=['Dose'],ascending=True)
percent_app
```

```python
# Plot

# Prepare the figure from the data
fig = percent_app.plot(kind = 'bar', 
                       figsize = (16, 10), 
                       width = 0.5, 
                       edgecolor = 'g', 
                       color = ['deepskyblue','mediumblue','blue','dodgerblue','skyblue','powderblue' ])


# Draw each group
for i in fig .patches:
    fig.text(i.get_x() + 0.0, 
             i.get_height() + 0.3, 
             str(round((i.get_height()), 2)), 
             fontsize = 15, 
             color = 'black',
             rotation = 0
             )

# Write the title and print
plt.title("Priority Group Comorbidity vs Application", fontsize = 15)
plt.xticks(rotation=0)
plt.show()
```

#### 10) And what about the manufacturers? Is there one brand of vaccine more frequent than the others in our sample?

```python
manufacturers = df['vacina_fabricante_nome'].value_counts()
fig, ax = plt.subplots(figsize=(15, 5))
plt.title('Applications per brand',fontsize=25)
manufacturers.plot(kind='bar', color = ['maroon'])
plt.xticks(rotation=45)
plt.show()
```

We can observe that Pfizer is ther has the highest share in our sample, followed by Sinovac/Butantan in second place.
